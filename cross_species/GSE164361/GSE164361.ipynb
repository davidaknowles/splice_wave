```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import re
import sys
```

```python
files = list(Path(".").glob("*.bed.gz"))
```

```python
def get_meta(m): 
    return [ m.group(i) for i in [1,2] ]

matches = [ 
    get_meta(re.search(r"GSE164361_([a-z]+)_(\w+)_ATAC_idr\.output\.bed\.gz", str(f)))
    for f in files ]
sample_meta = pd.DataFrame(matches, columns = ["species", "tissue"])
sample_meta["file"] = files
sample_meta["species"] = sample_meta["species"].map({
    "rat" : "rattus_norvegicus", 
    "human" : "homo_sapiens"
})
sample_meta["filename"] = files
sample_meta["original"] = files
sample_meta["assay"] = "ATAC"
sample_meta.head()
```

```python
meta = sample_meta[["tissue","assay","species","filename","original"]]
meta.to_csv("meta.tsv", sep="\t", index=False)
```
