```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from pathlib import Path
import re
import sys
```

```python
files = list(Path(".").glob("*.gz"))
```

```python
def get_meta(m): 
    return [ m.group(i) for i in [1,2] ]

matches = []
for f in files:
    matches.append( 
        get_meta(re.search(r"GSE243256_([hpf0-9]+)_(\w+).narrow_peak.gz", str(f)))
    )
sample_meta = pd.DataFrame(matches, columns = ["stage", "tissue"])
sample_meta["filename"] = files
sample_meta["original"] = files
sample_meta["assay"] = "ATAC"
sample_meta["species"] = "danio_rerio"


sample_meta
```

```python
meta = sample_meta[["tissue","assay","species","filename","original"]]
meta.to_csv("meta.tsv", sep="\t", index=False)
```
